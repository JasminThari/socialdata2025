```python
import pandas as pd 
import re
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
```

```python
# Load the data
df_prior2018 = pd.read_csv('../data/Police_Department_Incident_Reports__Historical_2003_to_May_2018_20250211.csv')
df_after2018 = pd.read_csv('../data/Police_Department_Incident_Reports__2018_to_Present_20250211.csv')
```

```python
df_prior2018['Date'] = pd.to_datetime(df_prior2018['Date'])
df_prior2018 = df_prior2018.sort_values(by='Date').reset_index(drop=True)

df_after2018['Incident Date'] = pd.to_datetime(df_after2018['Incident Date'])
df_after2018 = df_after2018.loc[df_after2018['Incident Date'] < '2025-01-01'] # using only complete years 
df_after2018 = df_after2018.sort_values(by='Incident Date').reset_index(drop=True)
```

```python
# Rename columns to match and filter out columns that are irrelevant 
df_prior2018.rename(columns={'IncidntNum': 'IncidentNumber', 
                             'Incident Code': 'IncidentCode',
                             'Category': 'IncidentCategory', 
                             'Descript': 'IncidentDescription', 
                             'DayOfWeek': 'IncidentDayOfWeek', 
                             'Date': 'IncidentDate', 
                             'Time': 'IncidentTime', 
                             'X': 'Longitude',
                             'Y': 'Latitude',
                             'location': 'Location', 
                             'PdDistrict': 'PoliceDistrict', 
                             'Neighborhoods 2': 'Neighborhoods',
                             'Current Police Districts 2 2': 'CurrentPoliceDistricts',
                             'Current Supervisor Districts 2 2': 'CurrentSupervisorDistricts',
                             'Analysis Neighborhoods 2 2': 'AnalysisNeighborhood',}, 
                    inplace=True)
df_after2018.rename(columns={'Incident Number': 'IncidentNumber',
                             'Incident Code': 'IncidentCode',
                            'Incident Category': 'IncidentCategory',
                            'Incident Day of Week': 'IncidentDayOfWeek',
                            'Incident Date': 'IncidentDate',
                            'Incident Time': 'IncidentTime',
                            'Incident Description': 'IncidentDescription',
                            'Police District': 'PoliceDistrict',
                            'Current Police Districts': 'CurrentPoliceDistricts',
                            'Current Supervisor Districts': 'CurrentSupervisorDistricts',
                            'Analysis Neighborhood': 'AnalysisNeighborhood',
                            'Point': 'Location', 
                            }, inplace=True)

df_prior2018_filtered_cols = df_prior2018[['IncidentNumber', 'IncidentCode', 'IncidentCategory', 'IncidentDescription', 'IncidentDayOfWeek', 
                                     'IncidentDate', 'IncidentTime', 'Resolution', 'Longitude', 'Latitude', 'Location', 
                                     'PoliceDistrict', 'PdId', 'Neighborhoods', 'CurrentPoliceDistricts', 'CurrentSupervisorDistricts',
                                     'AnalysisNeighborhood']]

df_after2018_filtered_cols = df_after2018[['IncidentNumber', 'IncidentCode', 'IncidentCategory', 'IncidentDescription', 'IncidentDayOfWeek', 
                                     'IncidentDate', 'IncidentTime', 'Resolution', 'Longitude', 'Latitude', 'Location', 
                                     'PoliceDistrict', 'Neighborhoods', 'CurrentPoliceDistricts', 'CurrentSupervisorDistricts',
                                     'AnalysisNeighborhood']]
```

```python
def capitalize_words(s):
    return re.sub(r'\b\w+\b', lambda match: match.group(0).capitalize(), s.lower())

df_prior2018_filtered_cols.loc[:, 'IncidentCategory'] = df_prior2018_filtered_cols['IncidentCategory'].apply(capitalize_words)

df_prior2018_filtered_cols.dropna(subset=['PoliceDistrict'], inplace=True)
df_prior2018_filtered_cols.loc[:, 'PoliceDistrict'] = df_prior2018_filtered_cols['PoliceDistrict'].apply(capitalize_words)
```

```python
# print(df_prior2018_filtered_cols['IncidentCategory'].unique())
# print(df_after2018_filtered_cols['IncidentCategory'].unique())
```

```python
category_mapping = {
    'Assault':'Assault', 
    'Trespass': 'Trespass',
    'Drug Violation': 'Drug/Narcotic',
    'Drug Offense': 'Drug/Narcotic',
    'Larceny Theft': 'Larceny/Theft',
    'Malicious Mischief': 'Vandalism',  
    'Motor Vehicle Theft': 'Vehicle Theft',
    'Motor Vehicle Theft?': 'Vehicle Theft',
    'Weapons Carrying Etc': 'Weapon Laws',
    'Weapons Offense': 'Weapon Laws',
    'Weapons Offence': 'Weapon Laws',
    'Liquor Laws': 'Drunkenness',
    'Vehicle Impounded': 'Recovered Vehicle', 
    'Human Trafficking, Commercial Sex Acts': 'Prostitution', 
    'Lost Property': 'Non-Criminal' , 
    'Civil Sidewalks':'Non-Criminal', 
    'Traffic Collision':'Non-Criminal', 
    'Human Trafficking (A), Commercial Sex Acts': 'Prostitution', 
    'Suspicious': 'Suspicious Occ', 
    'Fire Report':'Non-Criminal', 
    'Warrant':'Warrants', 
    'Forgery And Counterfeiting':'Forgery/Counterfeiting', 
    'Case Closure':'Non-Criminal', 
    'Vehicle Misplaced':'Recovered Vehicle', 
    'Rape':'Sex Offenses, Forcible', 
    'Courtesy Report':'Non-Criminal', 
    'Sex Offense':'Sex Offenses, Non Forcible', 
}
```

```python
df_after2018_filtered_cols.dropna(subset=['IncidentCategory'], inplace=True)
df_after2018_filtered_cols.loc[:, 'IncidentCategory'] = df_after2018_filtered_cols['IncidentCategory'].apply(lambda x: category_mapping[x] if x in category_mapping else x)
```

```python
# return categories not in both sets
other_categories = set(df_prior2018_filtered_cols['IncidentCategory'].unique()) ^ set(df_after2018_filtered_cols['IncidentCategory'].unique())
df_after2018_filtered_cols.loc[:, 'IncidentCategory'] = df_after2018_filtered_cols['IncidentCategory'].apply(lambda x: 'Other Offerenses' if x in other_categories else x)
df_prior2018_filtered_cols.loc[:, 'IncidentCategory'] = df_prior2018_filtered_cols['IncidentCategory'].apply(lambda x: 'Other Offerenses' if x in other_categories else x)
```

```python
set(df_prior2018_filtered_cols['PoliceDistrict'].unique()) ^ set(df_after2018_filtered_cols['PoliceDistrict'].unique())
```

```python
df_after2018_filtered_cols = df_after2018_filtered_cols.query('PoliceDistrict != "Out of SF"')
```

```python
df_all = pd.concat([df_prior2018_filtered_cols, df_after2018_filtered_cols], axis=0)
df_all.reset_index(drop=True, inplace=True)
```

```python
df_all_focus = df_all.query('IncidentCategory in @focuscrimes').reset_index(drop=True)
df_all_focus['Year'] = df_all_focus['IncidentDate'].dt.year
```

```python
import matplotlib.pyplot as plt

# Group data by year and category, then count occurrences
df_grouped = df_all_focus.groupby(['Year', 'IncidentCategory']).size().unstack(fill_value=0)

# Plot settings
df_grouped.plot(kind='bar', figsize=(20, 10), stacked=True)

# Titles and labels
plt.title('Crimes by Category Per Year', fontsize=24, pad=20)
plt.xlabel('Year', fontsize=18, labelpad=15)
plt.ylabel('Crime Count', fontsize=18, labelpad=15)

# Customize tick labels
plt.xticks(rotation=45, fontsize=14)
plt.yticks(fontsize=14)

# Grid and tight layout for better appearance
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group data by year and category, then count occurrences
df_grouped = df_all_focus.groupby(['Year', 'IncidentCategory']).size().unstack(fill_value=0)

# Plot settings
categories = df_grouped.columns
num_categories = len(categories)
cols = 3  # Number of columns for subplots
rows = (num_categories // cols) + (num_categories % cols > 0)

fig, axes = plt.subplots(rows, cols, figsize=(20, rows * 5))

for i, category in enumerate(categories):
    row, col = divmod(i, cols)
    ax = axes[row, col]
    
    # Plot each category in a separate subplot
    df_grouped[category].plot(kind='bar', ax=ax, color='skyblue', edgecolor='black')
    
    # Customize each subplot
    ax.set_title(category, fontsize=14)
    ax.set_xlabel('Year', fontsize=12)
    ax.set_ylabel('Crime Count', fontsize=12)
    ax.tick_params(axis='x', rotation=45)

    # Grid for each subplot
    ax.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout and remove empty subplots
plt.tight_layout()
for j in range(i + 1, rows * cols):
    fig.delaxes(axes.flat[j])

plt.suptitle('Crimes by Category Per Year', fontsize=24, y=1.02)
plt.show()
```
